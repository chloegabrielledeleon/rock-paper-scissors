```python
import random
import time
from IPython.display import clear_output

#CHLOE'S PART
def getMove():
  """
  WRITTEN BY: Chloe

  Get valid input from player. Returns player move once its valid

  ARGS: none

  RETURNS: Move that user inputs
  """
  isValidMove = False
  while isValidMove == False:
    move = input("Please enter rock, paper, or scissors: ").strip().lower()
    if (move == "rock") or (move == "paper") or (move == "scissors"):
      isValidMove = True
    else:
      print("Double check spelling before submitting.")

  return move

def whoWon(playerMove, AImove):
  """
  WRITTEN BY: Chloe

  Determines basic game logic. Rock beats scissors, paper beats rock, and scissors beats paper.

  ARGS: playerMove(rock, paper, scissors) and AImove(rock, paper, scissors)

  RETURNS: winner - either "AI", "player", none(if its a tie)
  """
  winner = "none" # set a default value
  if (playerMove == "rock") and (AImove == "scissors"):
    winner = "player"
  if (playerMove == "rock") and (AImove == "paper"):
    winner = "AI"
  if (playerMove == "paper") and (AImove == "rock"):
    winner = "player"
  if (playerMove == "paper") and (AImove == "scissors"):
    winner = "AI"
  if (playerMove == "scissors") and (AImove == "rock"):
    winner = "AI"
  if (playerMove == "scissors") and (AImove == "paper"):
    winner = "player"
  if (playerMove == AImove):
    winner = "none"

  return winner

def randomAI():
  """
  WRITTEN BY: Chloe

  Generates a random int 1 to 3, inclusive. 1 is rock, 2 is paper, 3 is scissors. Function returns the move associated with the number

  ARGS: none

  RETURNS: AImove - rock, paper, or scissors
  """
  moveNum = random.randint(1,3)
  if moveNum == 1:
    AImove = "rock"
  elif moveNum == 2:
    AImove = "paper"
  elif moveNum == 3:
    AImove = "scissors"

  return AImove

def patternAI():
  """
  WRITTEN BY: Chloe

  Pattern AI (looks for player patterns in last 3-5 moves). If there has been less than 3 moves, calls the randomAI() and returns a move using that.

  ARGS: none

  RETURNS: AImove - rock, paper, or scissors
  """
  global playerMoveList


  # if player has played less that 3 moves
  if len(playerMoveList) < 3:
    AImove = randomAI()

  # if player has played 3 moves
  elif len(playerMoveList) < 4:
    patternCommMove = mostCommonPlayerChoice(playerMoveList[-3:])
    if patternCommMove == "rock":
      AImove = "paper"
    elif patternCommMove == "paper":
      AImove = "scissors"
    elif patternCommMove == "scissors":
      AImove = "rock"

    else:
      AImove = randomAI()

  # if player has played 4 moves
  elif len(playerMoveList) < 5:
    patternCommMove = mostCommonPlayerChoice(playerMoveList[-4:])
    if patternCommMove == "rock":
      AImove = "paper"
    elif patternCommMove == "paper":
      AImove = "scissors"
    elif patternCommMove == "scissors":
      AImove = "rock"

    else:
      AImove = randomAI()

  # if player has played 5 or move moves
  else:
    patternCommMove = mostCommonPlayerChoice(playerMoveList[-5:])
    if patternCommMove == "rock":
      AImove = "paper"
    elif patternCommMove == "paper":
      AImove = "scissors"
    elif patternCommMove == "scissors":
      AImove = "rock"

    else:
      AImove = randomAI()

  return AImove

def counterAI():
  """
  WRITTEN BY: Chloe

  Takes player's most common move and returns the move that would beat it.
  If there is no common move, the AI returns a random move.
  Similar to patternAI, but uses all past moves. If no moves have been played yet or there is no common move, this section will chose a move instead using randomAI.

  ARGS: none

  RETURNS: AImove - rock, paper, or scissors
  """
  commMove = mostCommonPlayerChoice(playerMoveList)
  AImove = None
  if commMove == "rock":
    AImove = "paper"
  if commMove == "paper":
    AImove = "scissors"
  if commMove == "scissors":
    AImove = "rock"

  if commMove == "none":
    AImove = randomAI()

  return AImove


def getGameType():
  """
  WRITTEN BY: Chloe

  Meant to be called at the beginning. Asks the player if they want to play againt the randomAI, patternAI, or counterAI

  ARGS: none

  RETURNS: gameType - 'random', 'pattern', or 'counter'
  """
  isValidType = False
  while isValidType == False:
    gameType = input("Which of the following AIs do you want to play against: random, pattern, or counter? ").strip().lower()
    if (gameType == "random") or (gameType == "pattern") or (gameType == "counter"):
      isValidType = True
    else:
      print("Double check spelling before submitting.")

  return gameType

def whatAIPlays(gameType):
  """
  WRITTEN BY: Chloe

  Calls the AI player method corresponding to what gameType the user selected.

  ARGS: gameType, as determined by getGameType()

  RETURNS: AImove

  """
  if gameType == "random":
    AImove = randomAI()
  elif gameType == "pattern":
    AImove = patternAI()
  elif gameType == "counter":
    AImove = counterAI()

  return AImove

# STEPHEN'S PART
"""Student C: Statistics & Analysis - Track win/loss/tie for each throw type - Calculate win percentage - Track current
streak (wins/losses) - Longest win streak - Most common player choice - Most common AI choice - Head-to-head results
by throw type"""


playerRockWins = playerPaperWins = playerScissorsWins = 0
AIRockWins = AIPaperWins = AIScissorsWins = 0
playerRockLosses = playerPaperLosses = playerScissorsLosses = 0
AIRockLosses = AIPaperLosses = AIScissorsLosses = 0
rockTie = paperTie = scissorsTie = 0

def winType(playerMove, AIMove, winner):
  """
  WRITTEN BY: Stephen

  Determines what choice led to a win, loss, or tie.

  ARGS: playerMove - move chosen by the player, AIMove - move chosen by the AI, winner - the winner of the round

  RETURNS: The amount of wins, losses, ties, by each move
  """
  #global playerWinList, AIWinList, tieList
  global playerRockWins, playerPaperWins, playerScissorsWins
  global AIRockWins, AIPaperWins, AIScissorsWins
  global playerRockLosses, playerPaperLosses, playerScissorsLosses
  global AIRockLosses, AIPaperLosses, AIScissorsLosses
  global rockTie, paperTie, scissorsTie
  #Player wins
  if winner == 'player':
      if playerMove == 'rock':
            playerRockWins += 1
      elif playerMove == 'paper':
            playerPaperWins += 1
      elif playerMove == 'scissors':
            playerScissorsWins += 1

  #Player losses
  if winner == 'AI':
    if playerMove == 'rock':
            playerRockLosses += 1
    elif playerMove == 'paper':
            playerPaperLosses += 1
    elif playerMove == 'scissors':
            playerScissorsLosses += 1

   #AI wins
  elif winner == 'AI':
      if AIMove == 'rock':
            AIRockWins += 1
      elif AIMove == 'paper':
            AIPaperWins += 1
      elif AIMove == 'scissors':
            AIScissorsWins += 1

  #AI losses
  if winner == 'player':
    if AIMove == 'rock':
            AIRockLosses += 1
    elif AIMove == 'paper':
            AIPaperLosses += 1
    elif AIMove == 'scissors':
            AIScissorsLosses += 1

  #Ties
  elif winner == 'none':
      if playerMove == 'rock':
            rockTie += 1
      elif playerMove == 'paper':
            paperTie += 1
      elif playerMove == 'scissors':
            scissorsTie += 1

  return playerRockWins, playerPaperWins, playerScissorsWins, AIRockWins, AIPaperWins, AIScissorsWins, \
  playerRockLosses, playerPaperLosses, playerScissorsLosses, AIRockLosses, AIPaperLosses, AIScissorsLosses, \
  rockTie, paperTie, scissorsTie



totalGamesPlayed = 0
def gameCounter():
  """
  WRITTEN BY: Stephen

  Tracks the total amount of games played.

  ARGS: none

  RETURNS: gameCount - the number of total rounds played
  """
  global totalGamesPlayed
  totalGamesPlayed += 1
  return totalGamesPlayed


def winPercentage(AIWins, playerWins):
  """
  WRITTEN BY: Stephen

  Divides the amount of wins by the total amount of games to determine the win percentage.

  ARGS: AIWins - the number of AI wins, playerWINS - the number of player wins

  RETURNS: the winpercentages for both AI and the player
  """
  global totalGamesPlayed
  AIWinPercentage = AIWins / totalGamesPlayed
  playerWinPercentage = playerWins / totalGamesPlayed

  return AIWinPercentage, playerWinPercentage

playerWinStreak = 0
def currentStreak(winner):
  """
  WRITTEN BY: Stephen

  Keeps track of how many times in a row the player has won.

  ARGS: winner - who won the round

  RETURNS: playerWinStreak - keeps track of how many times in a row the player has won
  """
  if winner == 'player':
    global playerWinStreak
    playerWinStreak += 1
  return playerWinStreak

playerLongestWinStreak = 0
def longestWinStreak(playerWinStreak):
  """
  WRITTEN BY: Stephen

  Determines the longest amount of games the player has won in a row.

  ARGS: playerWinStreak - how many times the player has been winning in a row

  RETURNS: playerLongestWinStreak - the player's longest winstreak
  """
  global playerLongestWinStreak
  if playerWinStreak > playerLongestWinStreak:
    playerLongestWinStreak = playerWinStreak
  return playerLongestWinStreak


playerMoveList = []
def recordPlayerMoveList(playerMove):
  """
  WRITTEN BY: STEPHEN

  Keeps track of all moves made by the player.

  ARGS: playerMove - what move the player chose

  RETURNS: playerMoveList - a list of all moves made by the player
  """

  global playerMoveList
  if playerMove == 'rock':
    playerMoveList.append('rock')
  if playerMove == 'paper':
    playerMoveList.append('paper')
  if playerMove == 'scissors':
    playerMoveList.append('scissors')

  return playerMoveList

def mostCommonPlayerChoice(playerMoveList):
  """
  WRITTEN BY: STEPHEN

  Determines the most common player choice based on the amount of times the player has thrown rock, paper, or scissors.

  ARGS: playerMoveList - list of all player moves

  RETURNS: mostCommonPlayerMove - the most common player move
  """

  rockCount = playerMoveList.count("rock")
  paperCount = playerMoveList.count("paper")
  scissorsCount = playerMoveList.count("scissors")

  if rockCount > paperCount and rockCount > scissorsCount:
    mostCommonPlayerMove = "rock"
  elif paperCount > rockCount and paperCount > scissorsCount:
    mostCommonPlayerMove = "paper"
  elif scissorsCount > paperCount and scissorsCount> rockCount:
    mostCommonPlayerMove = "scissors"
  else:
    mostCommonPlayerMove = None
  return mostCommonPlayerMove



AIMoveList = []
def mostCommonAIChoice(AIMove):
  """
  WRITTEN BY: STEPHEN

  Keeps track of the AI's moves and then determines the most common move.

  ARGS: AIMove - the move made by AI

  RETURNS: mostCommonAIMove - the most common AI move

  Keeps track of the AI's most common choice."""

  if AIMove == 'rock':
    AIMoveList.append('rock')
  if AIMove == 'paper':
    AIMoveList.append('paper')
  if AIMove == 'scissors':
    AIMoveList.append('scissors')

  rockCount = AIMoveList.count("rock")
  paperCount = AIMoveList.count("paper")
  scissorsCount = AIMoveList.count("scissors")

  if rockCount > paperCount and rockCount > scissorsCount:
    mostCommonAIMove = "rock"
  elif paperCount > rockCount and paperCount > scissorsCount:
    mostCommonAIMove = "paper"
  elif scissorsCount > paperCount and scissorsCount > rockCount:
    mostCommonAIMove = "scissors"
  else:
    mostCommonAIMove = None
  return mostCommonAIMove


rockPlayerResultsList = []
rockAIResultsList = []
paperPlayerResultsList = []
paperAIResultsList = []
scissorsPlayerResultsList = []
scissorsAIResultsList = []
playerRockResults = []
playerPaperResults = []
playerScissorsResults = []
AIRockResults = []
AIPaperResults = []
AIScissorsResults = []

def headToHeadResults(results):
  #results is a list of playermove, aimove, who won
  """
  WRITTEN BY: STEPHEN

  How many times has the player and AI won/lost/tied with each throw type.

  ARGS: results - a list of the move made by the player, AI, and who won

  RETURNS: a list of the player or AI wins/losses/ties for each move type

  """


  if results[0] == 'rock' and results[2] == 'player':
    rockPlayerResultsList.append("win")
    rockAIResultsList.append("loss")
  elif results[0] == 'rock' and results[2] == 'AI':
    rockPlayerResultsList.append("loss")
    rockAIResultsList.append("win")
  elif results[0] == 'rock' and results[2] == 'none':
    rockPlayerResultsList.append("tie")
    rockAIResultsList.append("tie")


  if results[0] == 'paper' and results[2] == 'player':
    paperPlayerResultsList.append("win")
    paperAIResultsList.append('loss')
  elif results[0] == 'paper' and results[2] == 'AI':
    paperPlayerResultsList.append("loss")
    paperAIResultsList.append('win')
  elif results[0] == 'paper' and results[2] == 'none':
    paperPlayerResultsList.append("tie")
    paperAIResultsList.append('tie')


  if results[0] == 'scissors' and results[2] == 'player':
    scissorsPlayerResultsList.append("win")
    scissorsAIResultsList.append('loss')
  elif results[0] == 'scissors' and results[2] == 'AI':
    scissorsPlayerResultsList.append("loss")
    scissorsAIResultsList.append('win')
  elif results[0] == 'scissors' and results[2] == 'none':
    scissorsPlayerResultsList.append("tie")
    scissorsAIResultsList.append('tie')


#when player threw rock, they won this many times, lost this many times, tied thismany times
  playerRockResults.append([
      rockPlayerResultsList.count("win"),
      rockPlayerResultsList.count("loss"),
      rockPlayerResultsList.count("tie")])

  global lastPlayerRockResults
  lastPlayerRockResults = playerRockResults.pop()

  playerPaperResults.append([
      paperPlayerResultsList.count("win"),
      paperPlayerResultsList.count("loss"),
      paperPlayerResultsList.count("tie")])

  global lastPlayerPaperResults
  lastPlayerPaperResults = playerPaperResults.pop()

  playerScissorsResults.append([
      scissorsPlayerResultsList.count("win"),
      scissorsPlayerResultsList.count("loss"),
      scissorsPlayerResultsList.count("tie")])

  global lastPlayerScissorsResults
  lastPlayerScissorsResults = playerScissorsResults.pop()

  AIRockResults.append([
      rockAIResultsList.count("win"),
      rockAIResultsList.count("loss"),
      rockAIResultsList.count("tie")])

  global lastAIRockResults
  lastAIRockResults = AIRockResults.pop()

  AIPaperResults.append([paperAIResultsList.count("win"),
                         paperAIResultsList.count("loss"),
                         paperAIResultsList.count("tie")])

  global lastAIPaperResults
  lastAIPaperResults = AIPaperResults.pop()

  AIScissorsResults.append([
      scissorsAIResultsList.count("win"),
      scissorsAIResultsList.count("loss"),
      scissorsAIResultsList.count("tie")])

  global lastAIScissorsResults
  lastAIScissorsResults = AIScissorsResults.pop()


  return lastPlayerRockResults, lastPlayerPaperResults, lastPlayerScissorsResults, lastAIRockResults, lastAIPaperResults, lastAIScissorsResults
  #to get the win/loss/tie for each type, win = playerRockResults[0], loss = win = playerRockResults[1], tie = playerRockResults[2]

# ABIGAIL'S PART
"""
Student B: Tournament System - Best-of-3, best-of-5, best-of-7 formats - Round tracking (display current round,
scores) - Match winner determination - Series winner determination - Ability to play multiple series
"""

# set up tournament system
def tournamentstyle():
  """
  Written by Abigail

  prompts the user with how many rounds of a series they will play

  Args: none

  Returns: the number of rounds in a series, either 3, 5, or 7
  """
  rounds = int(input("Would you like to play to best of 3, 5, or 7? "))
  while (not(rounds == 3 or rounds == 5 or rounds ==7)):
    # edge case protection
    print("Please enter a valid input")
    rounds = int(input("Would you like to play to the best of 3, 5, or 7?"))
  return rounds

def playGame(gameType, rounds, currentSeries, totalSeries):
  """
  Written by Abigail with help from Chloe

  playGame() runs through all of the move of one rounds of rock paper scissors

  ARGS: gameType, rounds, currentSeries, totalSeries

  RETURNS: moveMoveWinList (move of AI and player and the winner in a list), rounds

  """
  global playerWins
  global playerWinStreak
  playerWins = 0
  AIWins = 0
  playerWinStreak = 0

  for i in range(rounds):

    print()
    #prints the round and series numbers
    print(f"------ Series {currentSeries} of {totalSeries} ------")
    print(f"----- Game {i+1} of {rounds} games -----")

    playerMove = getMove()


    AImove = whatAIPlays(gameType)
    mostCommonAIChoice(AImove)
    playerMoveList = recordPlayerMoveList(playerMove)

    time.sleep(1)

    winner = whoWon(playerMove, AImove)
    gameCounter()
    winType(playerMove, AImove, winner)

    # determine the winner of each round
    print(f"AI move: {AImove}")

    if winner == "player":
      print("You won this round")
      playerWins += 1 # if the player wins a round, add a win to a player win counter
      streak = currentStreak('player')
      longestWinStreak(streak)
    elif winner == "AI":
      print("You lost this round")
      AIWins += 1 # add win to AI win counter
      playerWinStreak = 0 #restart the longest win streak for loss or tie
    else:
      print("this round was a tie")
      playerWinStreak = 0


    time.sleep(3)
    moveMoveWinList = [playerMove, AImove, winner]
    headToHeadResults(moveMoveWinList) # Call moveMoveWinList here so that head-to-head results are kept track of in each round

    # clear_output(wait=True)

  return moveMoveWinList, rounds, playerWins, AIWins


def serieswinner(rounds):
  """
  Written by Abigail

  determine and print the winner of a set of three, five, or seven games

  ARGS: rounds- number of rounds in a series

  RETURNS: nothing

  """
  if playerWins > rounds/2: # if the user has the majority of wins
    print("You won this set of games")
    time.sleep(2)
    # clear_output()
  elif playerWins < rounds /2:
    print("The AI won this set of games")
    time.sleep(2)
    # clear_output()
  else:
    print("This series was a tie")
    time.sleep(2)
    # clear_output()

# multiple series
def multipleseries():
  """
  Written by abigail, chloe, and stephen

  this is the main method of out project. there is a loop to play multiple series (loop of playGame()). after all
  of the series are played, a statistics report of all games played is printed

  ARGS: none

  RETURNS: none
  """
  seriesnum = int(input("How many series do you want to play? "))
  gameType = getGameType()
  rounds = tournamentstyle()

  #print("series num: ", seriesnum)

  totalPlayerWins = 0
  totalAIWins = 0
  for i in range(seriesnum):
    print("Starting in 3 seconds...")
    time.sleep(3)
    # clear_output(wait=True)

    print("start game")
    moveMoveWinList, rounds, playerWins, AIWins = playGame(gameType, rounds, i + 1, seriesnum)
    totalPlayerWins += playerWins
    totalAIWins += AIWins #keeps track of total wins so that it does not resent every series
    serieswinner(rounds)
    time.sleep(2)
    # clear_output(wait=True)

  playerWinPercent, AIWinPercent = winPercentage(totalAIWins, totalPlayerWins)

  clear_output(wait=True)

  print("\n")
  print("----- STATISTICS REPORT -----")
  print("Games played: " + str(totalGamesPlayed))
  print(f"Your win percentage: {playerWinPercent:.2%}")
  print(f"AI win percentage: {AIWinPercent:.2%}")
  print(f"Your longest win streak: {playerLongestWinStreak}")
  print(f"Your most common choice: {mostCommonPlayerChoice(playerMoveList)}")
  print(f"AI's most common choice: {mostCommonAIChoice(AIMoveList)}")
  print(f"When you threw rock, you won {lastPlayerRockResults[0]} time(s), lost {lastPlayerRockResults[1]} time(s), and tied {lastPlayerRockResults[2]} time(s).")
  print(f"When you threw paper, you won {lastPlayerPaperResults[0]} time(s), lost {lastPlayerPaperResults[1]} time(s), and tied {lastPlayerPaperResults[2]} time(s).")
  print(f"When you threw scissors, you won {lastPlayerScissorsResults[0]} time(s), lost {lastPlayerScissorsResults[1]} time(s), and tied {lastPlayerScissorsResults[2]} time(s).")
  print(f"When AI threw rock, AI won {AIRockWins} time(s), lost {AIRockLosses} time(s), and tied {rockTie} time(s).")
  print(f"When AI threw paper, AI won {AIPaperWins} time(s), lost {AIPaperLosses} time(s), and tied {paperTie} time(s).")
  print(f"When AI threw scissors, AI won {AIScissorsWins} time(s), lost {AIScissorsLosses} time(s), and tied {scissorsTie} time(s).")

multipleseries()
```
